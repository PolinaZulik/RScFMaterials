```python
limocc = 10
ttypes = ['big5', 'fpi']
kwset = set()
resd = {}
for ttype in ttypes:
  foo1 = './drive/MyDrive/rscf/rusidiolect_data/corrs/' + ttype +'/Пол/'
  fns = [x for x in os.listdir(foo1) if x.endswith('.tsv') and (x.startswith('interp') or ('_lex_' in x))]
  for fn in fns:
    df = pd.read_csv(foo1+fn, sep = '\t', index_col=0, header = [0,1])
    if ('Unnamed: 0', 'Unnamed: 0') in df.columns:
      df.drop(('Unnamed: 0', 'Unnamed: 0'), axis=1, inplace=True)
    df.columns = df.columns.map(lambda x: (x[0].rstrip('1234567890.'), x[1].rstrip('1234567890.')))
    kws = df.columns.levels[0]
    for kw in kws:
      resd[kw] = resd.get(kw, [])
      data = df[kw].dropna()
      data = data[data.Occ >= limocc]
      data.rename({'Word':'Word/Dim', 'Dim':'Word/Dim'}, axis=1, inplace = True)
      data['Text'] = fn
      resd[kw].append(data)
    kwset.update(kws)
```

```python
resfoo = './drive/MyDrive/rscf/rusidiolect_data/corrs/corrs_output/Пол/'
for kw in kwset:
  resdf = pd.concat(resd['я_PRON'])
  resdf[resdf.columns[3:6]] = resdf[resdf.columns[3:6]].astype(int)
  resdf['T-Stat'] = resdf['T-Stat'].round(2)
  resdf['P-value'] = resdf['P-value'].round(5)
  resdf['P-Bonferroni'] = resdf['P-Bonferroni'].round(1)
  resdf.to_csv(resfoo+kw+'.tsv', index = False)
```
