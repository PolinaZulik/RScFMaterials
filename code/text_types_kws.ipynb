```python
import os 
import pandas as pd
```

```python
def getfilewords(fn='556_1_описание.csv'):
  df = pd.read_csv(foo + fn, index_col=0)
  kws = set(['_'.join(x) for x in zip(df.Lemma, df.POS)])
  return list(kws)
```

```python
foo = './drive/MyDrive/rscf/rusidiolect_data/processed_spacy/fpi_1045/'
fns = os.listdir(foo)
fns1 = [x[:-4].split('_') for x in fns]
```

```python
typevcs = {}
typens = ['sex', 'type']
for i, j in enumerate([2, -1]):
  types = pd.Series([x[j] for x in fns1])
  vc = types.value_counts()
  typevcs[typens[i]] = vc
typevcs
```

```python
res = {}
for k, vc in typevcs.items():
  if k == 'sex':
    nameind = 2
  elif k == 'type':
    nameind = 4
  for idx in vc.index:
    print(idx, vc.loc[idx])
    allkws = []
    for i, fn in enumerate(fns):
      if fns1[i][nameind] == idx:
        allkws += getfilewords(fn)
      if i%50 == 0:
        print(i, 'done.') 
    len(allkws)
    dc = pd.Series(allkws).value_counts()
    dc = dc.rename_axis(str(idx) +' / '+str(vc.loc[idx]))
    res[idx] = dc
#  dc.to_csv('./drive/MyDrive/rscf/rusidiolect_data/processed_spacy/kws_fpi_' + idx +'.csv')
```
