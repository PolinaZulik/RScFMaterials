```python
def corr_any_trait(fullfn, texttype, alldf, trait = 'Текст: Возраст автора', binary = False):
  df1 = pd.read_csv(fullfn, sep = '\t', index_col=0, header = 0, skiprows = [0], usecols = range(301))
  df2 = df1.loc[(df1!=0).any(axis=1)]
  #df2 = df1
  df3 = df2.div(df2.sum(axis=1), axis=0).fillna(0.)
  keys = ['Author', 'Age', 'Gender', 'NText', 'TypeText']
  ss = [s.split('_') for s in df3.index]
  d = [{keys[i]: s[i] for i in range(5)} for s in ss]
  ids = pd.DataFrame(d, index = df3.index, columns = keys)
  ttype = ids[ids.TypeText == texttype].index
  #print(len(ttype))
  df4 = df3.loc[ttype]
  wocc = df4.shape[0]
  if wocc < 2:
    return None
  #print(ttype, df4)
  infotrait = alldf.loc[df4.index][trait]
  if binary:
    corrname = 'T-Stat'
    datagroup = {}
    woccsb = {}
    for bval in binary:
      datagroup[bval] = df4.loc[infotrait.apply(lambda x: x.startswith(bval))]
      woccsb[bval] = df4.loc[infotrait.apply(lambda x: x.startswith(bval))].shape[0]
    #print(datagroup[binary[0]])
    resdata = stats.ttest_ind(datagroup[binary[0]], datagroup[binary[1]], equal_var=False) #Welch test!
    #print(resdata)
    if type(resdata[0]) == type(np.nan):
      return None
    res = pd.DataFrame(resdata, columns = datagroup[binary[0]].columns).transpose()
    for bval in binary:
      res['Occ(' + bval+')'] = woccsb[bval]
    #res
    #print(res)
    #print(woccs.shape)
  else:
    corrname = 'Pearson C'
    res = df4.corrwith(infotrait, method = stats.pearsonr).transpose()
  res['Occ'] = wocc
  res = res.reset_index()
  res.rename({'index': 'Dim', 0: corrname, 1: 'P-value'}, axis=1, inplace = True)
  if 'P-value' in res.columns:
    res['P-Bonferroni'] = res['P-value']*wocc
    res.sort_values('P-value', ascending = True, inplace = True, ignore_index=True)
  else:
    #print(res, woccs)
    res = None
  return res
def procfile_any_trait(fullfn, texttype, trait, alldf, lim = None, binary = False):
  res = corr_any_trait(fullfn, texttype, alldf, trait, binary = binary)
  if type(res) == type(None):
    return res
  if lim:
    #print('Lim')
    res = res[res['P-value'] < lim]
    res.reset_index(inplace=True, drop = True)
  return res
```

```python
rfoo = './drive/MyDrive/rscf/rusidiolect_data/corrs/big5/'
resfoos = os.listdir(rfoo)
for fulltrait in ['Автор: Пол']:
  trait = fulltrait.split(': ')[1]
  if not os.path.isdir(rfoo+trait):
    print(trait, resfoos)
    os.mkdir(rfoo+trait)
  print(trait)

  fo = foo+'w2v_contexts/big5_536/'
  fns = [fn for fn in os.listdir(fo) if fn.endswith('.tsv')]
  for ttype in ['повествование', 'описание', 'рассуждение']:
    print('-',ttype)
    kwsres = [procfile_any_trait(fo+fn, ttype, trait = fulltrait, alldf = adf, lim = 0.05, binary = ['м', 'ж']) for fn in fns]
    signres, kwcols = [], []
    for i,kw in enumerate(kwsres):
      if (type(kw) != type(None)) and kw.shape[0]:
        signres.append(kw)
        kwcols.append(fns[i])
    if len(kwcols):
      res = pd.concat(signres, axis=1, keys = [fn[:-4] for fn in kwcols])
      res.to_csv('./drive/MyDrive/rscf/rusidiolect_data/corrs/big5/'+trait+'/big5_w2v_' +ttype+'.tsv', sep = '\t')
```
