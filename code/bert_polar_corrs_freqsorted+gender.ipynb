```python
def getcontexts(fn):
  df1 = pd.read_csv(fn, sep = '\t', index_col=0, encoding = 'utf-8', encoding_errors='ignore')
  dfpresent = df1[df1.apply(lambda x: x != 0.).any(axis=1)]
  return dfpresent
```

```python
def getWinverse(df):
  ma = df.dirvec.apply(lambda x:x.flatten()).tolist()
  W = np.matrix(ma)
  W_inverse = linalg.pinv(np.transpose(W))
  return W_inverse
```

```python
def getpolarmatrix(freqtype = 'harm-freq', first = 100):
  matrfoo = './drive/MyDrive/rscf/interp_polar/'
  df2 = pd.read_pickle(matrfoo+'direction_vectors_1try_freqs.pkl')
  df2 = df2.sort_values(by = freqtype, ascending = False)
  df2 = df2[:first]
  W_inverse = getWinverse(df2)
  polarnames = df2.scale.apply(lambda x: x.replace('\t', ' - '))
  return W_inverse, polarnames
```

```python
def getpolarperson(fn):
  dfpresent = getcontexts(fn)
  W_inverse, polarnames = getpolarmatrix()
  polarperson = np.matmul(W_inverse, dfpresent.to_numpy().T)
  ppdf = pd.DataFrame(polarperson.T, columns = polarnames, index = dfpresent.index)

  return ppdf
```

```python
respath = './drive/MyDrive/rscf/rusidiolect_data/corrs/bert_polar_100highestfreq/krusk_spear/'
fn = fns[0]
psyfold = 'big5_536'
def getcorrs(fn, psyfold = 'big5_536'):
  ppdf = getpolarperson(foo+psyfold+'/'+fn)
  psy = getpsy(psyfold)
  psy1 = psy.loc[ppdf.index.intersection(psy.index)]
  Np = psy1.shape[0]
  print(fn, '\t', Np, 'participants')
  res = {}
  for i, c in enumerate(psy1.columns):
    meth = lambda x, y: signtest(x, y, method = spearmanr)
    if c == 'gender':
      meth = lambda x, y: signtest(x, y, method = kruskal)
    res[c] = ppdf.corrwith(psy1[c], method = meth, axis=0)
  allres = pd.concat(res)
  allres.dropna(how = 'all', axis=1, inplace = True)
  Rs = allres.query("ilevel_1 == 0").droplevel(1).T
  Ps = allres.query("ilevel_1 == 1").droplevel(1).T
  resfn = fn.split('_')[0] + '_' + str(Np) +'.xlsx'
  with pd.ExcelWriter(respath + psyfold +'/' + resfn) as writer:
    Rs.to_excel(writer, sheet_name = 'Stat')
    Ps.to_excel(writer, sheet_name = 'P-value')
```
