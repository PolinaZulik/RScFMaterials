```python
import csv
import numpy as np
import os
import csv
from deeppavlov.core.common.file import read_json
from deeppavlov import build_model, configs
```

```python
import glob
way = glob.glob('./drive/MyDrive/rusidiolect_data/processed_spacy/lex_contexts/fpi_1045/*')
kwlist = []  # сюда добавляем сами ключевые слова
for i in way:
  i = i.replace('./drive/MyDrive/rusidiolect_data/processed_spacy/lex_contexts/fpi_1045/', '')
  i = i.replace('.tsv', '')
  kwlist.append(i)
```

```python
bert_config = read_json(configs.embedder.bert_embedder)
bert_config['metadata']['variables']['BERT_PATH'] = 'rubert-base-cased-conversational'  # config.json --> bert_config.json

m = build_model(bert_config)
```

```python
new = []
for kw in kwlist:
  if 'й' in kw:
    print(kw)
    kw = kw.replace('й', 'й')
    new.append(kw)
  else:
    new.append(kw)
kwlist = new
```

```python
for i in kwlist:
  print(i)
```

```python
ch_files = []
for i in files:
  if i not in ch_files:
    ch_files.append(i)

print(len(ch_files))
```

```python
for kw in kwlist[18:]:

  # f = open('./drive/MyDrive/rusidiolect_data/processed_spacy/bert_contexts/fpi_1045/'+kw+'.tsv', 'a')
  f = open('fpi_1045/'+kw+'.tsv', 'a')

  first_line = [' ']

  for num in range(1, 769):
    first_line.append(str(num))
  kw_word = kw.split('_')[0]

  f.write('\t'.join(first_line) + '\n')  # записываем первую строку

  for x, y in zip(files, texts):
    row = [x]
    doc = nlp(y)  # передаём для токенизации текст y
    to_bert = []  # сюда будут попадать предложения с КС, которые попадут в модель

    for sentence in doc.sentences:  # в терминах Stanza (тот же SpaCy, только новее) 
      for word in sentence.words:
        if word.lemma == kw_word:  # если слово равно ключевому
          to_bert.append(sentence.text)  # предложение добавляем в список

    if len(to_bert) == 0:  # если длина нашего списка с предложениями равна 0, то весь вектор будет нулевым
      for num in range(1, 769):
        row.append(str(0))
      f.write('\t'.join(row) + '\n')
    else:
      try:
        bert_vectors = []
        tokens, token_embs, subtokens, subtoken_embs, sent_max_embs, sent_mean_embs, bert_pooler_outputs = m(to_bert)
        c = 0  # индекс токенизированных предложений tokens = [[...], [...], [...]], один подсписок -- одно предложение
        for sent_token in tokens:  # для каждого подсписка токенов предложения
          for word_token in sent_token:  # для каждого токена в предложении
            doc = nlp(word_token)  # в Stanza обрабатываем токен и получаем характеристики
            for sentence in doc.sentences:
              for word in sentence.words:
                if kw_word == word.lemma:
                  ind = sent_token.index(word_token)
                  bert_vectors.append(np.array(token_embs)[c][ind])
                  #print(sent_token)
                  #print(word_token)
                  #print(np.array(token_embs)[c][ind])
          c += 1
        final_bert_vector = sum(bert_vectors)/len(bert_vectors)
        for vec in final_bert_vector:
          row.append(str(vec))
        f.write('\t'.join(row) + '\n')
      except:
        for num in range(1, 769):
          row.append(str(0))
        f.write('\t'.join(row) + '\n')

  f.close()
```
