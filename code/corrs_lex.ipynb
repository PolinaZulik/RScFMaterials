```python
def corr_any_trait(fullfn, texttype, alldf, trait = 'Текст: Возраст автора', binary = False):
  df1 = pd.read_csv(fullfn, sep = '\t', index_col=0)
  df2 = df1.drop('count', axis=1)
  df3 = df2.div(df2.sum(axis=1), axis=0).fillna(0.)
  keys = ['Author', 'Age', 'Gender', 'NText', 'TypeText']
  ss = [s.split('_') for s in df3.index]
  d = [{keys[i]: s[i] for i in range(5)} for s in ss]
  ids = pd.DataFrame(d, index = df3.index, columns = keys)
  ttype = ids[ids.TypeText == texttype].index
  woccs = (df2.loc[ttype] >0).sum()
  df4 = df3.loc[ttype]
  #print(alldf)
  infotrait = alldf.loc[df4.index][trait]
  if binary:
    corrname = 'T-Stat'
    datagroup = {}
    woccsb = {}
    for bval in binary:
      datagroup[bval] = df4.loc[infotrait.apply(lambda x: x.startswith(bval))]
      woccsb[bval] = (df2.loc[ttype].loc[infotrait.apply(lambda x: x.startswith(bval))] >0).sum()
    #print(datagroup[binary[0]])
    resdata = stats.ttest_ind(datagroup[binary[0]], datagroup[binary[1]], equal_var=False) #Welch test!
    #print(resdata)
    if type(resdata[0]) == type(np.nan):
      return None
    res = pd.DataFrame(resdata, columns = datagroup[binary[0]].columns).transpose()
    for bval in binary:
      res['Occ(' + bval+')'] = woccsb[bval].astype(int)
    #res
    #print(res)
    #print(woccs.shape)
  else:
    corrname = 'Pearson C'
    res = df4.corrwith(infotrait, method = stats.pearsonr).transpose()
  res['Occ'] = woccs.astype(int)
  res = res.reset_index()
  res.rename({'index': 'Word', 0: corrname, 1: 'P-value'}, axis=1, inplace = True)
  if 'P-value' in res.columns:
    res['P-Bonferroni'] = res['P-value']*woccs.shape[0]
    res.sort_values('P-value', ascending = True, inplace = True, ignore_index=True)
  else:
    #print(res, woccs)
    res = None
  return res
def procfile_any_trait(fullfn, texttype, trait, alldf, lim = None, binary = False):
  res = corr_any_trait(fullfn, texttype, alldf, trait, binary = binary)
  
  if type(res) == type(None):
    return res
  #print(res.shape, res)
  if lim:
    res = res[(res['Occ'] >= lim[0]) & (res['P-value'] < lim[1])]
    res.reset_index(inplace=True, drop = True)
  return res
```

```python
def corr_gender(df3, texttype):
  keys = ['Author', 'Age', 'Gender', 'NText', 'TypeText']
  ss = [s.split('_') for s in df3.index]
  d = [{keys[i]: s[i] for i in range(5)} for s in ss]
  ids = pd.DataFrame(d, index = df3.index, columns = keys)
  ttype = ids[ids.TypeText == texttype]
  male = ttype[ttype.Gender == 'м']
  female = ttype[ttype.Gender == 'ж']
  mdata = df3.loc[male.index]
  fdata = df3.loc[female.index]
  cols = mdata.columns
  res = [stats.ttest_ind(mdata[c], fdata[c]) for c in cols]
  rest = [[cols[i], x[0], x[1], x[1]*len(cols), sum(mdata[cols[i]] > 0), sum(fdata[cols[i]] > 0)] for i,x in enumerate(res)]
  gender_res = pd.DataFrame(rest, columns = ['Word', 'Stat', 'P-value', 'P-Bonferroni', 'Occ Male', 'Occ Fem'])
  gender_res['Occ'] = gender_res['Occ Male'] + gender_res['Occ Fem']
  gender_res.sort_values(['P-value', 'Occ'], inplace = True, ignore_index=True)
  return gender_res[gender_res.columns[:-1]]

def procfile_gender(fullfn, texttype, lim = None):
  df1 = pd.read_csv(fullfn, sep = '\t', index_col=0)
  df2 = df1.drop('count', axis=1)
  df3 = df2.div(df2.sum(axis=1), axis=0).fillna(0.)
  res = corr_gender(df3, texttype = texttype)
  if lim:
    res = res[((res['Occ Male'] >= lim[0]) | (res['Occ Fem'] >= lim[0])) & (res['P-value'] < lim[1])]
    res.reset_index(inplace=True, drop = True)
  return res
```
